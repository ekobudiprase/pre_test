```python
import numpy as np
import pandas as pd
import statistics as st

churn = pd.read_csv('churn.csv', encoding='latin1')
churn.head()
```

```python
# 1. parameter apa yang memiliki korelasi paling tinggi kelas churn?
# jawab : parameter international_plan

churn.corr(method='pearson')
```

```python
#2. bagaimana statistik untuk number_customer_service_calls? (mean, modus, median)

print("mean : ", st.mean(churn['number_customer_service_calls']) )
print("modus : ", st.mode(churn['number_customer_service_calls']) )
print("median : ", st.median(churn['number_customer_service_calls']) )
print("std dev : ", st.pstdev(churn['number_customer_service_calls']) )

churn['number_customer_service_calls'].value_counts()
```

```python
# 3. Bagaimanakah pengaruh international_plan thd churn customer
# jawab : dari data di bawah, pelanggan yg tidak ada international_plan cenderung akan menjadi pelanggan churn


churn.groupby(['class','international_plan'], as_index=False).count()
```

```python
# 4. Bagaimanakah pengaruh number_customer_service_calls thd churn customer
# jawab : dari data di bawah, semakin pelanggan melakukan customer service call maka akan menekan kenaikan customer churn


churn.groupby(['class','number_customer_service_calls'], as_index=False).count()
```

```python
# 5. Bagaimanakah pengaruh voice_mail_plan & international_plan thd churn customer
# jawab : dari data di bawah, pelanggan yang tidak ada plan voice mail & international plan maka cenderung lebih riskan menjadi churn customer


churn.groupby(['class','voice_mail_plan','international_plan'], as_index=False).count()
```

```python
#2. Bagaimanakah pengaruh number_vmail_messages thd churn customer
#jawab : dari data di bawah, semakin besar number_vmail_messages cenderung makin besar churn customer

churn.groupby(['class','number_vmail_messages'], as_index=False).count()
```
